```python
import numpy as np

#Plotting age, balance, and housing loan status

data_yes = data[data['housing']=='yes']
data_no = data[data['housing']=='no']

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(15,5))

ax1.scatter(data_no['age'],data_no['balance'],s=5,color='blue')
ax1.set_xlabel("age", fontsize = 20)
ax1.set_ylabel("balance", fontsize = 20)
ax1.set_title('Housing Loan = no')
ax1.set_xlim([10, 100])

ax2.scatter(data_yes['age'],data_yes['balance'],s=5,color='red')
ax2.set_xlabel("age", fontsize = 20)
ax2.set_ylabel("balance", fontsize = 20)
ax2.set_title('Housing Loan = yes')
ax2.set_xlim([10, 100])
```

```python
import numpy as np

#Plotting age, balance, and personal loan status

data_yes = data[data['loan']=='yes']
data_no = data[data['loan']=='no']

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(15,5))

ax1.scatter(data_no['age'],data_no['balance'],s=5,color='purple')
ax1.set_xlabel("age", fontsize = 20)
ax1.set_ylabel("balance", fontsize = 20)
ax1.set_title('Personal Loan = no')
ax1.set_xlim([10, 100])

ax2.scatter(data_yes['age'],data_yes['balance'],s=5,color='orange')
ax2.set_xlabel("age", fontsize = 20)
ax2.set_ylabel("balance", fontsize = 20)
ax2.set_title('Personal Loan = yes')
ax2.set_xlim([10, 100])
```

```python
#Comparing average balance per age, per job

data3 = data.groupby(['age','job']).balance.mean()
data3 = data3.unstack(level='job')
data3.columns.name = 'balance'
data3 = data3.fillna(0)
data3.plot(subplots=True, figsize=(15, 15), sharex=True, sharey=True); plt.legend(loc='best')
plt.xlabel("Age")
plt.ylabel("Average Balance")
```
